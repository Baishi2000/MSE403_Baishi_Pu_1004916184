### <span style="color: red"> Task 2:</span> Who has the most runner-up (i.e., second place) finishes in the histroy of NHL? Create a table with the teams and their runner-up finishes. The table must contain the the following columns:

-Team

-Total number of runner-up finishes

-Years [list of the years they finished as a runner-up]

You have to use Jupyter for the practice and produce a notebook with good readability.

### <span style="color: blue"> Solution:</span>

```python
import pandas as pd
import numpy as np

data = pd.read_excel('Stanely_cup_winners_by_year.xlsx')

runup_df = pd.DataFrame({'Team':[], 'Runner up finish':[], 'Runner up years':[]})

for i in range(len(data)):
    
    current_team = data['Runner up'][i]
    current_year = str(data['Year'][i])
    # Check every row. If exist in dict, then add 1, add year. If not, add name, add 1, add year
    
    if not current_team in runup_df['Team'].values: # If encountered a new runner-up
        new_row = [current_team, 1, [current_year]] # initialize, by making a new row from list
        runup_df.loc[len(runup_df)] = new_row # add this row to end of df
        
    else:
        #here is the method to locate a row with specific value is 
        # https://stackoverflow.com/questions/16683701/in-pandas-how-to-get-the-index-of-a-known-value
        j = runup_df.loc[runup_df['Team'] == current_team].index[0] # j is used as row index within runup_df
        
        # print(runup_df['Runner up finish'][j])
        
        runup_df.loc[j, 'Runner up finish'] += 1 # add 1 to its runner up finish
        runup_df.loc[j, 'Runner up years'].append(current_year) # append current_year to the Runner up years list
        
runup_df = runup_df.sort_values(by = ['Runner up finish'], ascending = False)
runup_df = runup_df.set_index(np.arange(0, len(runup_df)))
runup_df
```

### <span style="color: red"> Task 3:</span> With the help of matminer, 1. Retrive the 'steel_strength' dataset into a dataframe and clean your dataset (i.e., do some basic data exploration and analysis work on your data. For instance, check the statistics of the dataset, check for NaNs, drop duplicated data entries etc. It is alway the first thing to do before you start working on a dataset.) if needed; 2. Featurize the compositions of the materials; and 3. Perform correlation analysis and drop feature(s) with a coefficient greater than abs(0.8) from your dataframe.

### <span style="color: blue"> Solution:</span>

```python
import matminer
from matminer.datasets import load_dataset
#matminer.datasets.get_available_datasets() #you can check what datasets are available with this line

strength = load_dataset('steel_strength')
strength = strength.dropna(axis = 0)
strength = strength.set_index(np.arange(0, len(strength)))

from matminer.featurizers.conversions import StrToComposition 
#make a column of matminer composition & structure
strength = StrToComposition().featurize_dataframe(strength, 'formula')
# strength = CompositionToStructureFromMP().featurize_dataframe(strength, 'composition')

#add two more features
from matminer.featurizers.composition.alloy import YangSolidSolution
from matminer.featurizers.composition.packing import AtomicPackingEfficiency

strength = YangSolidSolution().featurize_dataframe(strength, 'composition')
strength = AtomicPackingEfficiency().featurize_dataframe(strength, 'composition')
```

```python
df = strength
df.head(5)
```

```python
# create copy of only features
features = df.iloc[:, -7:]

#below method is from: 
#https://stackoverflow.com/questions/29294983/how-to-calculate-correlation-between-all-columns-and-remove-highly-correlated-on

corr_matrix = features.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.8
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]

# Drop features 
features.drop(to_drop, axis=1, inplace=True)

df = df.iloc[:, :-7].join(features)

df
```
